```python
from sklearn.preprocessing import minmax_scale, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
# à¸—à¸³ normalization à¸”à¹‰à¸§à¸¢à¸§à¸´à¸˜à¸µ standard scaler
scalar = StandardScaler()
new_penguin = df[:]
numerical_col = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm','body_mass_g']
new_penguin[numerical_col] = scalar.fit_transform(new_penguin[numerical_col])
new_penguin
```

```python
new_penguin = pd.get_dummies(new_penguin, columns=['island'], prefix='island')
new_penguin
```

```python
new_penguin = pd.get_dummies(new_penguin, columns=['sex'], prefix='sex')
new_penguin
```

```python
X = new_penguin.drop(columns='species')
y = new_penguin['species']
```

```python
x_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state=10)
```

```python
knn = KNeighborsClassifier()
```

```python
param_grid = {'n_neighbors' : list(range(1,31)),
            'weights' : ['uniform', 'distance'],
            'metric' : ['euclidean', 'manhattan']
}
```

```python
grid_search = GridSearchCV(estimator= knn, param_grid = param_grid, cv = 5, n_jobs = -1 , scoring = 'accuracy')
```

```python
grid_search.fit(x_train,y_train)
```

```python
print("Best parameters", grid_search.best_params_)
print("Best cross-validation score", grid_search.best_score_)
```

```python
# best_estimator à¸„à¸·à¸­ knn à¸—à¸µà¹ˆà¹€à¸—à¸£à¸™à¹à¸¥à¹‰à¸§à¸à¸±à¸šà¸žà¸²à¸£à¸²à¸¡à¸´à¹€à¸•à¸­à¸£à¹Œà¸—à¸µà¹ˆà¸”à¸µà¸—à¸µà¹ˆà¸ªà¸¸à¸”
best_knn = grid_search.best_estimator_
y_pred = best_knn.predict(X_test)
```

```python
print('Accuracy', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
